```python
# create an array with help of linspace
#scholl time
#Write 5 numbers equally spaced between 10 and 20.

# We know:
# Start = 10
# End = 20
# Total numbers = 5
# Step 1: Find the gap
# If there are 5 numbers, there are 4 gaps between them.
# 10  _  _  _  _ 20   ?
# 20 - 10  % 4 = 2.5
#  10 + 2.5 + 12.5 + 2.5  + 15+ 2.5 + 17.5 + 2.5 = 20 ( ans )
# Same way linspace work

a = np.linspace(10,20,5)
print('linspace : ',a)
```

Summary of joining of nd arrays:
--------------------------------------------------
1).concatenate()==>Join a sequence of arrays along an existing axis.  <br>
2).stack()==>Join a sequence of arrays along a new axis.<br>
3).vstack()==>Stack arrays in sequence vertically according to first axis(axis-0).<br>
4).hstack()==>Stack arrays in sequence horizontally according to 2nd axis(axis-1).<br>
5)dstack()==>Stack arrays in sequence depth wise according to 2rd axis(axis-2).<br>


In 2-D array axis-0 means rows(axis:-2) <br>
In 2-D array axis-1 means columns(axis:-1)

```python
import numpy as np
import matplotlib.pyplot as plt


marks=np.array([80,70,90,60,100])
print(np.var(marks))
plt.plot(marks, marker="o")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.title("Student Marks")
plt.grid(True)
plt.show()
```

Note : <b> var </b>  How far the numbers are spread from the average. <br>

```python
hours = np.array([2,3,4,5])
marks = np.array([50,65,70,85])

print(np.corrcoef(hours,marks))
x = np.corrcoef(hours,marks)
plt.hist(x)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X=np.array([[2],[4],[6],[8]])
y=np.array([50,60,70,85])

model=LinearRegression()

model.fit(X,y)

print(model.predict([[5]]))
```

Topic	Simple Meaning <br>
Mean	Average <br>
Median	Middle value <br>
Mode	Most repeated value <br>
Range	Highest − Lowest <br>
Variance	Average squared distance from the mean <br>
Standard Deviation	Typical distance from the mean <br>
Probability	Chance of an event <br>
Normal Distribution	Bell-shaped pattern where most values are near the center<br>
Correlation	How two variables move together<br>
Regression	Predict a number from other data

<strong> Now let's see which statistic answers which question. </strong>

If you want to know...	   Use  
What is the average mark?	Mean    <br>
What is the middle mark?	Median   <br>
Which mark appears most?	Mode     <br>
How far apart are the highest and lowest marks?	Range  <br>
Are students' marks very different from each other?	Variance  <br>
On average, how far is each mark from the mean?	Standard Deviation  <br>
What is the chance of getting Head when tossing a coin?	Probability <br>
Is the data bell-shaped?	Normal Distribution  <br>
Does studying more increase marks?	Correlation <br>
Can I predict marks from study hours?	Regression
