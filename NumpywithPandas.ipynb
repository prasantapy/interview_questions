```python
# create an array with help of linspace
#scholl time
#Write 5 numbers equally spaced between 10 and 20.

# We know:
# Start = 10
# End = 20
# Total numbers = 5
# Step 1: Find the gap
# If there are 5 numbers, there are 4 gaps between them.
# 10  _  _  _  _ 20   ?
# 20 - 10  % 4 = 2.5
#  10 + 2.5 + 12.5 + 2.5  + 15+ 2.5 + 17.5 + 2.5 = 20 ( ans )
# Same way linspace work

a = np.linspace(10,20,5)
print('linspace : ',a)
```

Summary of joining of nd arrays:
--------------------------------------------------
1).concatenate()==>Join a sequence of arrays along an existing axis.  <br>
2).stack()==>Join a sequence of arrays along a new axis.<br>
3).vstack()==>Stack arrays in sequence vertically according to first axis(axis-0).<br>
4).hstack()==>Stack arrays in sequence horizontally according to 2nd axis(axis-1).<br>
5)dstack()==>Stack arrays in sequence depth wise according to 2rd axis(axis-2).<br>


In 2-D array axis-0 means rows(axis:-2) <br>
In 2-D array axis-1 means columns(axis:-1)
