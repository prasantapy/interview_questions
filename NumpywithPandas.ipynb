5)dstack()==>Stack arrays in sequence depth wise according to 2rd axis(axis-2).<br>


In 2-D array axis-0 means rows(axis:-2) <br>

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.arange(1, 11)

Y = []
for i in A:
    if i % 2 == 0:
        Y.append(i)
    else:
        Y.append(np.nan)   # Skip odd numbers

plt.title("Even Number Points")
plt.plot(A, Y, marker='o')
plt.xlabel("X Range")
plt.ylabel("Even Values")
plt.grid(True)
plt.show()
```

```python
# Sequence of activities of plot function
# ----------------------------------------------------------
# 1.Creation of figure object
# 2.Creation of plot/axes object
# 3.Darw x & y axis
# 4.Mark evenly spaced values on x-axis and y-axis(xticks and yticks)
# 5.Plot the data points
# 6.Connect these data points with line
# 7.Add title, xlabel and ylabel


A = np.arange(1,11,dtype=object)
new = []
for x in A:
  if x%2==0:
    new.append(x)
  else:
    new.append(np.nan)
plt.title('Grap of even and odd ',{'color':'red','backgroundcolor':'Yellow'})
plt.plot(A,new,marker='o',color='green')
plt.xlabel(' X List of number ' ,{'color':'red','backgroundcolor':'Yellow'})
plt.ylabel('Y list of number',{'color':'red','backgroundcolor':'Yellow'})
plt.grid(color='g',visible=True,lw=1,axis='x')
plt.minorticks_on()
plt.legend()
plt.show()
```
